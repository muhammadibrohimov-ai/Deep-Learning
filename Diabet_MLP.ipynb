# Diabet kasalligini Perseptron orqali aniqlash

```python
# PyTorch kutubxonasi

import torch
import torch.nn as nn # neyron-tarmoq
import torch.optim as optim # optimizatsiya algoritmi

import matplotlib.pyplot as plt
%matplotlib inline

X = torch.tensor([              # tensor oddiy arrayning torch holatida foydalanish (GPU uchun)
                    [0.2, 0.3],
                    [0.6, 0.8],
                    [0.4, 0.5],
                    [0.7, 0.9]
                ], dtype=torch.float32)

y = torch.tensor([[0], [1], [0], [1]], dtype = torch.float32)
```

```python
model = nn.Sequential(
    nn.Linear(2, 4), # 2 ta ma'lumot qabul qiladi , 4 neyronga beradi
    nn.ReLU(),
    nn.Linear(4, 1),
    nn.Sigmoid()
)
```

```python
loss = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)

epoches = 100

losses = []

for epoch in range(epoches):
    y_predict = model(X)
    loss_value = loss(y_predict, y)
    losses.append(loss_value.item())

    optimizer.zero_grad()
    loss_value.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f"Epoxa: {epoch} - loss: {loss_value.item()}")
```

```python
plt.plot(losses)
plt.xlabel("Eopxa")
plt.ylabel("Loss")
plt.show()
```
